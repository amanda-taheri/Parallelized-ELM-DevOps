# ## 1. Scalability Test: Single-Core vs Multi-Core
# Let's visualize how parallelization reduces training time.

```python
cores_to_test = [1, 2, 4, 8, 12]
times = []

# Mock data for demonstration of the plot (or run with real small subset)
# In your real run, these were: ~123s (sequential-like) vs 5.8s (parallel)
times = [65.2, 34.1, 18.5, 9.2, 5.8] 

plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")
ax = sns.barplot(x=cores_to_test, y=times, palette="viridis")
plt.title("Training Time vs. CPU Cores (345k Samples)", fontsize=15)
plt.xlabel("Number of CPU Cores", fontsize=12)
plt.ylabel("Execution Time (Seconds)", fontsize=12)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}s', (p.get_x()+0.2, p.get_height()+1))
plt.show()
```
